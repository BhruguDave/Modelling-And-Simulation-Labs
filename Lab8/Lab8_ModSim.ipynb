```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

# 2D Random Walk

```python
def randomWalkPoints(n):

  #np.random.seed(0)
  pointsx = []
  pointsy = []
  pointsx.append(0)
  pointsy.append(0)
  x = 0
  y = 0
  for i in range(n):
    rand = 10000*np.random.uniform(0,10)//1
    if rand%2 == 0:
      x += 1
    else:
      x -= 1

    rand = 10000*np.random.uniform(0,10)//1
    if rand%2 == 0:
      y += 1
    else:
      y -= 1

    pointsx.append(x)
    pointsy.append(y)

  return pointsx, pointsy
```

```python
def animateWalk(x, y):
  tempx = []
  tempy = []
  
  for i in range(0, len(x)):
    tempx.append(x[i])
    tempy.append(y[i])
    plt.figure()
    plt.plot(tempx, tempy)
    plt.grid()
    plt.savefig('frame2 '+ str(i) + '.png')
```

```python
def meanRandomWalkDistance(n, sims):
  sum = 0
  for i in range(sims):
    x,y = randomWalkPoints(n)
    dist = math.sqrt((x[0] - x[len(x)-1])**2 + (y[0] - y[len(y)-1])**2)
    sum += dist

  return sum/sims
```

```python
x, y = randomWalkPoints(10)
plt.plot(x,y)
plt.grid()
plt.savefig('pehli.png')
```

```python
animateWalk(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

def animate(i):
    ax1.clear()
    ax1.plot(x[:i+1], y[:i+1], label = 'cases')
    ax1.set_xlim(np.min(x), np.max(x))
    ax1.set_ylim(np.min(y), np.max(y))

ani = animation.FuncAnimation(fig, animate, frames=11, interval=1000, blit=False, repeat=False)
ani
```

```python
n = np.arange(1, 101, 1)
sims = 100
mean = []
for i in n:
  mean.append(meanRandomWalkDistance(i,sims))

plt.figure()
plt.plot(n, mean)
plt.grid()
plt.xlabel('Number of Steps')
plt.ylabel('Mean Distance Travelled')
plt.savefig('mean-pehlo.png')
```

### BIASED

```python
def brandomWalkPoints(n,b):

  #np.random.seed(0)
  pointsx = []
  pointsy = []
  pointsx.append(0)
  pointsy.append(0)
  x = 0
  y = 0
  for i in range(n):
    rand = np.random.uniform(0,100)//1
    if rand > 100*b:
      x += 1
    else:
      x -= 1

    rand = np.random.uniform(0,100)//1
    if rand > 100*b:
      y += 1
    else:
      y -= 1

    pointsx.append(x)
    pointsy.append(y)

  return pointsx, pointsy
```

```python
bias = [0.1, 0.2, 0.4, 0.8]
n = 10
for b in bias:
  x,y = brandomWalkPoints(n, b)
  plt.figure()
  plt.plot(x,y)
  plt.grid()
  plt.title('Biased Random Walk p='+str(b)+'.png')
```

```python
bias = [0.1, 0.2, 0.4, 0.5, 0.8]
leg = []
for b in bias:
  leg.append('p='+str(b))
n = np.arange(1,101,1)
sims = 100
plt.figure()
for b in bias:
  mean = []
  sum = 0
  for i in n:
    for j in range(sims):
      x,y = brandomWalkPoints(i,b)
      sum += math.sqrt((x[0] - x[len(x)-1])**2 + (y[0] - y[len(y)-1])**2)
    mean.append(sum/sims)

  plt.plot(n, mean)

plt.grid()
plt.legend(leg)
plt.xlabel('Number of steps')
plt.ylabel('Mean distance travelled')
plt.savefig('bias-mean.png')
```
