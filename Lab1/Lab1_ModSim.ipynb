```python
# Importing Required Libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
def computation():
  Na = np.zeros(total_iter)
  Nb = np.zeros(total_iter)
  Nc = np.zeros(total_iter)

  Na[0] = N0

  for i in range(1,total_iter):
    Na[i] = Na[i-1] - h*ra*Na[i-1]
    Nb[i] = Nb[i-1] + h*ra*Na[i-1] - h*rb*Nb[i-1]
    Nc[i] = Nc[i-1] + h*rb*Nb[i-1]

  return Na, Nb, Nc
```

```python
def plot_masses():
  plt.figure(1)
  plt.plot(np.linspace(T_start, T_end, total_iter), Na, 'red')
  plt.plot(np.linspace(T_start, T_end, total_iter), Nb, 'orange')
  plt.plot(np.linspace(T_start, T_end, total_iter), Nc, 'green')
  plt.xlabel('time')
  plt.ylabel('mass of substance')
  plt.legend(['Mass of A', 'Mass of B', 'Mass of C'])
```

```python
def total_disintegration():
  total_dis = np.zeros(total_iter)
  for i in range(total_iter):
    total_dis[i] = ra*Na[i] + rb*Nb[i];

  plt.figure(2)
  plt.plot(np.linspace(T_start, T_end, total_iter), total_dis);
  plt.xlabel('time')
  plt.ylabel('total disintegrations')

  #return max(total_dis)
```

```python
def check_equilibrium():
  plt.figure(3)
  plt.plot(np.linspace(T_start, T_end, total_iter), Nb/Na);
  plt.xlabel('time')
  plt.ylabel('Mb/Ma')
```

```python
# Test Case - 1
# Rb >> Ra

N0 = 1000     # Initial Quantitity of A
T_end = 100
T_start = 0
h = 0.001     # Stepsize
ra = 0.001    # Deacay rate of A
rb = 0.6      # Decay rate of B
total_iter = int((T_end - T_start)/h)

Na, Nb, Nc = computation()
plot_masses()
total_disintegration()
check_equilibrium()
```
