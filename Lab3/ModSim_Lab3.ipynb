```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

### Q1 - Constant Harvesting Without death in isolation 

```python
def const_harv(n0, t, dt, harv):

  t_arr = np.arange(0, t+dt, dt)
  n_arr = 0 * t_arr
  n_arr[0] = n0

  for i in range(1, len(t_arr)):
    n_arr[i] = n_arr[i-1] + dt * n_arr[i-1] * (1 - n_arr[i-1]) - dt * harv

  return t_arr, n_arr
```

```python
growth_rate = 1
carrying_capacity = 1000
initial_population = 20
total_time = 12
dt = 0.01
harvesting_rate = 1

n0 = initial_population/carrying_capacity
tau = total_time * growth_rate
dtau = dt * growth_rate
hbar = harvesting_rate/(growth_rate * carrying_capacity)

t_arr, n_arr = const_harv(n0, tau, dtau, hbar)
plt.figure()
plt.plot(t_arr/ growth_rate, n_arr * carrying_capacity)
plt.xlabel('time')
plt.ylabel('population')
plt.grid()
plt.savefig("q1.png")
```

### Q1 - Dynamic Harvasting without death in isolation 

```python
def dynamic_harv(n0, t, dt, epsilon):

  t_arr = np.arange(0, t+dt, dt)
  n_arr = 0 * t_arr

  n_arr[0] = n0
  for i in range(1, len(t_arr)):
    n_arr[i] = n_arr[i-1] + dt * n_arr[i-1] * (1 - n_arr[i-1]) - dt * epsilon * n_arr[i-1]

  return t_arr, n_arr
```

```python
growth_rate = 1
carrying_capacity = 1000
initial_population = 20
total_time = 15
dt = 0.01
dynamic_harvesting_constant = 0.001

n0 = initial_population/carrying_capacity
tau = total_time * growth_rate
dtau = dt * growth_rate

t_arr, n_arr = dynamic_harv(n0, tau, dtau, dynamic_harvesting_constant)
plt.figure()
plt.plot(t_arr/ growth_rate, n_arr * carrying_capacity)
plt.xlabel('time')
plt.ylabel('population')
plt.grid()
plt.savefig("q1_dy.png")
```

### Non Harvesting with death in isolation

```python
def pop_simulation(n0, nthresh, t, dt):

  t_arr = np.arange(0, t+dt, dt)
  n_arr = 0 * t_arr

  n_arr[0] = n0

  for i in range(1, len(t_arr)):
    n_arr[i] = n_arr[i-1] + dt * n_arr[i-1] * (1 - n_arr[i-1]) * (n_arr[i-1] - nthresh)

  return t_arr, n_arr
```

```python
growth_rate = 2
carrying_capacity = 1000
death_by_isolation = 400
initial_population = 401
total_time = 20
dt = 0.01

n0 = initial_population/carrying_capacity
tau = total_time * growth_rate
dtau = dt * growth_rate
nthresh = death_by_isolation/carrying_capacity

t_arr, n_arr = pop_simulation(n0, nthresh, tau, dtau)
plt.figure()
plt.plot(t_arr/ growth_rate, n_arr * carrying_capacity)
plt.xlabel('time')
plt.ylabel('population')
plt.grid()
plt.savefig("q2_normal_grow.png")
```

```python
growth_rate = 2
carrying_capacity = 1000
death_by_isolation = 400
initial_population = 399
total_time = 20
dt = 0.01

n0 = initial_population/carrying_capacity
tau = total_time * growth_rate
dtau = dt * growth_rate
nthresh = death_by_isolation/carrying_capacity

t_arr, n_arr = pop_simulation(n0, nthresh, tau, dtau)
plt.figure()
plt.plot(t_arr/ growth_rate, n_arr * carrying_capacity)
plt.xlabel('time')
plt.ylabel('population')
plt.grid()
plt.savefig("q2_normal_death.png")
```
